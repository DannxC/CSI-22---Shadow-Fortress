```python
from PIL import Image
import numpy as np
img = np.array(Image.open('map'+mapnum+'.png'))
current_map = [0, 0]
#if a value is greater than 240 then make it 255, else make it 0
for i in range(len(img)):
    for j in range(len(img[i])):
        if img[i][j][0] > 240:
            img[i][j][0] = 255
        else:
            img[i][j][0] = 0
        if img[i][j][1] > 240:
            img[i][j][1] = 255
        else:
            img[i][j][1] = 0
        if img[i][j][2] > 240:
            img[i][j][2] = 255
        else:
            img[i][j][2] = 0
        if img[i][j][3] > 240:
            img[i][j][3] = 255
        else:
            img[i][j][3] = 0
```

```python
def get_green_corners(pixel_e, direction):
    pixel = pixel_e
    first_green, second_green, third_green, fourth_green = [0,0], [0,0], [0,0], [0,0]
    corners = []
    if direction == 'horizontal_bottom':
        found = False
        while not found:
            if img[pixel[0]][pixel[1]][0] == 0 and img[pixel[0]][pixel[1]][1] == 255 and img[pixel[0]][pixel[1]][2] == 0:
                first_green = pixel.copy()
                found = True
            else:
                pixel[1] -= 1
        found = False
        pixel[1] += 1
        while not found:
            if img[pixel[0]][pixel[1]][0] == 0 and img[pixel[0]][pixel[1]][1] == 255 and img[pixel[0]][pixel[1]][2] == 0:
                third_green = pixel.copy()
                found = True
            else:
                pixel[1] += 1
        found = False
        pixel[0] -= 1
        while not found:
            if img[pixel[0]][pixel[1]][0] == 0 and img[pixel[0]][pixel[1]][1] == 255 and img[pixel[0]][pixel[1]][2] == 0:
                fourth_green = pixel.copy()
                found = True
            else:
                pixel[0] -= 1
        second_green = [fourth_green[0], first_green[1]]
        corners = [first_green, second_green, third_green, fourth_green]
    if direction == "horizontal_top":
        found = False
        while not found:
            if img[pixel[0]][pixel[1]][0] == 0 and img[pixel[0]][pixel[1]][1] == 255 and img[pixel[0]][pixel[1]][2] == 0:
                second_green = pixel.copy()
                found = True
            else:
                pixel[1] -= 1
        found = False
        pixel[1] += 1
        while not found:
            if img[pixel[0]][pixel[1]][0] == 0 and img[pixel[0]][pixel[1]][1] == 255 and img[pixel[0]][pixel[1]][2] == 0:
                fourth_green = pixel.copy()
                found = True
            else:
                pixel[1] += 1
        found = False
        pixel[0] += 1
        while not found:
            if img[pixel[0]][pixel[1]][0] == 0 and img[pixel[0]][pixel[1]][1] == 255 and img[pixel[0]][pixel[1]][2] == 0:
                third_green = pixel.copy()
                found = True
            else:
                pixel[0] += 1
        first_green = [third_green[0], second_green[1]]
        corners = [first_green, second_green, third_green, fourth_green]
    if direction == "vertical_left":
        found = False
        while not found:
            if img[pixel[0]][pixel[1]][0] == 0 and img[pixel[0]][pixel[1]][1] == 255 and img[pixel[0]][pixel[1]][2] == 0:
                second_green = pixel.copy()
                found = True
            else:
                pixel[0] -= 1
        found = False
        pixel[0] += 1
        while not found:
            if img[pixel[0]][pixel[1]][0] == 0 and img[pixel[0]][pixel[1]][1] == 255 and img[pixel[0]][pixel[1]][2] == 0:
                first_green = pixel.copy()
                found = True
            else:
                pixel[0] += 1
        found = False
        pixel[1] += 1
        while not found:
            if img[pixel[0]][pixel[1]][0] == 0 and img[pixel[0]][pixel[1]][1] == 255 and img[pixel[0]][pixel[1]][2] == 0:
                third_green = pixel.copy()
                found = True
            else:
                pixel[1] += 1
        fourth_green = [second_green[0], third_green[1]]
        corners = [first_green, second_green, third_green, fourth_green]
    if direction == "vertical_right":
        found = False
        while not found:
            if img[pixel[0]][pixel[1]][0] == 0 and img[pixel[0]][pixel[1]][1] == 255 and img[pixel[0]][pixel[1]][2] == 0:
                fourth_green = pixel.copy()
                found = True
            else:
                pixel[0] -= 1
        found = False
        pixel[0] += 1
        while not found:
            if img[pixel[0]][pixel[1]][0] == 0 and img[pixel[0]][pixel[1]][1] == 255 and img[pixel[0]][pixel[1]][2] == 0:
                third_green = pixel.copy()
                found = True
            else:
                pixel[0] += 1
        found = False
        pixel[1] -= 1
        while not found:
            if img[pixel[0]][pixel[1]][0] == 0 and img[pixel[0]][pixel[1]][1] == 255 and img[pixel[0]][pixel[1]][2] == 0:
                first_green = pixel.copy()
                found = True
            else:
                pixel[1] -= 1
        second_green = [fourth_green[0], first_green[1]]
        corners = [first_green, second_green, third_green, fourth_green]
    return corners
```

```python
def get_next(img, first_green, second_green, third_green, fourth_green, came_from, current_room):
    rooms = []
    #search for the top room
    for i in range(second_green[1]+1, fourth_green[1]):
        if img[second_green[0]+1][i][0] == 255 and img[second_green[0]+1][i][1] == 0 and img[second_green[0]+1][i][2] == 0:
            if came_from != 'top':
                fir_green, sec_green, thi_green, fou_green = get_green_corners([second_green[0]-1, i], "horizontal_bottom")
                rooms.extend(get_next(img, fir_green, sec_green, thi_green, fou_green, 'bottom', [current_room[0], current_room[1]+1]))
            break
    #search for the bottom room
    for i in range(first_green[1]+1, third_green[1]):
        if img[first_green[0]-1][i][0] == 255 and img[first_green[0]-1][i][1] == 0 and img[first_green[0]-1][i][2] == 0:
            if came_from != 'bottom':
                fir_green, sec_green, thi_green, fou_green = get_green_corners([first_green[0]+1, i], "horizontal_top")
                rooms.extend(get_next(img, fir_green, sec_green, thi_green, fou_green, 'top', [current_room[0], current_room[1]-1]))
            break
    #search for the left room
    for i in range(second_green[0]+1, first_green[0]):
        if img[i][second_green[1]+1][0] == 255 and img[i][second_green[1]+1][1] == 0 and img[i][second_green[1]+1][2] == 0:
            if came_from != 'left':
                fir_green, sec_green, thi_green, fou_green = get_green_corners([i, second_green[1]-1], "vertical_right")
                rooms.extend(get_next(img, fir_green, sec_green, thi_green, fou_green, 'right', [current_room[0]-1, current_room[1]]))
            break
    #search for the right room
    for i in range(fourth_green[0]+1, third_green[0]):
        if img[i][fourth_green[1]-1][0] == 255 and img[i][fourth_green[1]-1][1] == 0 and img[i][fourth_green[1]-1][2] == 0:
            if came_from != 'right':
                fir_green, sec_green, thi_green, fou_green = get_green_corners([i, fourth_green[1]+1], "vertical_left")
                rooms.extend(get_next(img, fir_green, sec_green, thi_green, fou_green, 'left', [current_room[0]+1, current_room[1]]))
            break
    rooms.append([current_room, first_green, second_green, third_green, fourth_green])
    return rooms
```

```python
with open('map'+mapnum+'.txt') as f:
    data = f.readlines()
mapa = []
for d in data:
    #remove newline
    d = d[:-1]
    d = d.split(',')
    #transform string to int
    if(d[0]!="room" and d[0]!="spikes" and d[0]!="spawn"): d = tuple(map(int, d))
    else : d = (d[0], int(d[1]), int(d[2]))
    mapa.append(d)
#create a dictionary with the rooms
rooms = {}
spikes = {}
spawn = {}
is_spike = False
is_spawn = False
current_room = (0,0)
for i in range(len(mapa)):
    if mapa[i][0] == "room":
        current_room = (mapa[i][1], mapa[i][2])
        rooms[current_room] = []
        is_spike = False
        is_spawn = False
    elif mapa[i][0] == "spikes":
        spikes[current_room] = []
        is_spike = True
        is_spawn = False
    elif mapa[i][0] == "spawn":
        is_spike = False
        is_spawn = True
        spawn[current_room] = []
    else:
        if is_spike:
            spikes[current_room].append(mapa[i])
        elif is_spawn:
            spawn[current_room].append(mapa[i])
        else:
            rooms[current_room].append(mapa[i])
for room in rooms:
    rooms[room].sort(key=lambda a: a[1], reverse=True)
for spike in spikes:
    spikes[spike].sort(key=lambda a: a[1], reverse=True)
for sp in spawn:
    spawn[sp].sort(key=lambda a: a[1], reverse=True)
```
